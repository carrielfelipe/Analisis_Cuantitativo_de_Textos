```python
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import networkx as nx
from sklearn.manifold import TSNE
from gensim.models import Word2Vec
from nltk.tokenize import sent_tokenize, word_tokenize


# Función para generar el gráfico de red
def generar_grafico_red(text, model, title, offset_text=10, xlim=None, ylim=None, zlim=None, seed=42):
    # Fijar las semillas para la reproducibilidad
    np.random.seed(seed)
    random.seed(seed)

    # Dividir el texto en oraciones
    sentences = sent_tokenize(text, language='spanish')
    
    # Tokenizar las oraciones en palabras
    tokenized_sentences = [word_tokenize(sentence, language='spanish') for sentence in sentences]
    
    # Obtener las palabras del texto y sus embeddings del modelo entrenado
    words_in_text = set([word for sentence in tokenized_sentences for word in sentence if word in model.wv])
    word_vectors = np.array([model.wv[word] for word in words_in_text])
    
    # Reducir a 3 dimensiones con t-SNE
    tsne = TSNE(n_components=3, perplexity=5, random_state=seed)
    reduced_vectors = tsne.fit_transform(word_vectors)

    # Verificar si 'jardín' está en el vocabulario del texto
    if 'jardín' in words_in_text:
        index_jardin = list(words_in_text).index('jardín')
        reduced_vector_jardin = reduced_vectors[index_jardin]
        print("Vector reducido para 'jardín':")
        print(reduced_vector_jardin)
    else:
        print("'jardín' no está en el vocabulario del texto.")

    # Crear un gráfico de red
    G = nx.Graph()

    # Agregar nodos y posiciones de las palabras del texto
    similarity_threshold = 0.2
    for i, word in enumerate(words_in_text):
        G.add_node(word, pos=(reduced_vectors[i, 0], reduced_vectors[i, 1], reduced_vectors[i, 2]))  # Posición 3D

    # Calcular similitudes y agregar bordes solo entre las palabras del texto
    words_in_text_list = list(words_in_text)
    for i, word in enumerate(words_in_text_list):
        for j in range(i + 1, len(words_in_text_list)):
            # Calcular la similitud coseno
            similarity = np.dot(word_vectors[i], word_vectors[j]) / (np.linalg.norm(word_vectors[i]) * np.linalg.norm(word_vectors[j]))
            if similarity > similarity_threshold:
                G.add_edge(word, words_in_text_list[j])

    # Dibujar el gráfico en 3D
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')

    # Eliminar el fondo gris del gráfico 3D
    ax.xaxis.pane.fill = False  # Fondo del eje X
    ax.yaxis.pane.fill = False  # Fondo del eje Y
    ax.zaxis.pane.fill = False  # Fondo del eje Z

    # Eliminar las líneas de los bordes 
    ax.xaxis.pane.set_edgecolor('w')  # Color de borde del eje X (w para blanco)
    ax.yaxis.pane.set_edgecolor('w')  # Color de borde del eje Y
    ax.zaxis.pane.set_edgecolor('w')  # Color de borde del eje Z

    # Desactivar las rejillas
    ax.grid(False)
    #ax.set_axis_off()
    # Extraer posiciones de los nodos
    pos = nx.get_node_attributes(G, 'pos')

    # Crear un colormap
    cmap = cm.get_cmap('viridis', 256)  # 256 colores para un gradiente suave

    # Obtener los colores para los nodos basados en la posición X de las palabras
    x_positions = [pos[word][0] for word in words_in_text]
    norm = plt.Normalize(min(x_positions), max(x_positions))  # Normalizar posiciones para el gradiente
    node_colors = [cmap(norm(x)) for x in x_positions]  # Colores para los nodos

    # Dibujar los nodos en 3D
    for word in words_in_text:
        x, y, z = pos[word]
        ax.scatter(x, y, z, color=cmap(norm(x)), s=500, edgecolors='black')  # Color de borde para mayor claridad
        ax.text(x, y, z + offset_text, word, fontsize=10, ha='center', va='bottom')  # Etiquetas desplazadas

    # Dibujar los bordes
    for edge in G.edges:
        x_vals = [pos[edge[0]][0], pos[edge[1]][0]]
        y_vals = [pos[edge[0]][1], pos[edge[1]][1]]
        z_vals = [pos[edge[0]][2], pos[edge[1]][2]]
        ax.plot(x_vals, y_vals, z_vals, color='gray', alpha=0.5)

    ax.view_init(elev=30, azim=60)  # Cambia los valores de 'elev' y 'azim' para diferentes vistas

    # Establecer límites del gráfico si se especifican
    if xlim is not None:
        ax.set_xlim(xlim)
    if ylim is not None:
        ax.set_ylim(ylim)
    if zlim is not None:
        ax.set_zlim(zlim)

    # Título y ajustes finales
    ax.set_title(f'{title}')
    ax.set_xlabel('Componente 1')
    ax.set_ylabel('Componente 2')
    ax.set_zlabel('Componente 3')
    plt.show()


# Texto del sujeto esquizofrénico con menor TTR
texto_esquizofrenico = 'El jardín está bien las flores están, el sol está, todo está. Las cosas están, el perro está, la mesa está No sé todo está ahí.'

# Texto del sujeto sano con mayor variedad de palabras
texto_sano = 'El jardín está hermoso. Las flores han florecido con colores brillantes, y las hojas de los árboles se mueven suavemente con la brisa. El césped está verde y el aire fresco. Se siente paz aquí.'

# Tercer texto más similar al texto sano
texto_tercero = 'El jardín está lleno de vida. Los pájaros cantan mientras las flores exhalan su dulce aroma. El aire puro y el sol radiante crean un ambiente de tranquilidad.'

# Nuevos textos similares al texto sano
texto_cuarto = 'El jardín es un refugio de calma. Las mariposas revolotean entre las flores vibrantes, y el suave murmullo del viento acompaña el canto de los pájaros. La naturaleza respira paz.'
texto_quinto = 'En el jardín, las flores despliegan sus colores vibrantes, creando un hermoso paisaje. El cielo azul se refleja en el agua del estanque, mientras las hojas susurran con el viento.'
texto_sexto = 'El jardín es un lugar mágico. Las flores brillan bajo la luz del sol, y el canto de los pájaros llena el aire con melodías. Cada rincón invita a disfrutar de la serenidad de la naturaleza.'
texto_septimo = 'En el jardín, los colores estallan en una sinfonía de belleza. Las flores bailan al ritmo de la brisa, y el aroma de la tierra fresca llena el ambiente de vitalidad y alegría.'
texto_octavo = 'El jardín florece con vida. Las flores en tonos cálidos ofrecen un festín visual, mientras el canto alegre de los pájaros complementa la armonía del entorno. Todo se siente en perfecta sintonía.'

# Combinar los textos en un solo conjunto
todos_los_textos = [texto_esquizofrenico, texto_sano, texto_tercero, texto_cuarto, texto_quinto, texto_sexto, texto_septimo, texto_octavo]
sentences_combined = []
for text in todos_los_textos:
    sentences_combined.extend(sent_tokenize(text, language='spanish'))

# Tokenizar las oraciones combinadas en palabras
tokenized_sentences_combined = [word_tokenize(sentence, language='spanish') for sentence in sentences_combined]

# Entrenar un modelo Word2Vec con los textos combinados
model_combined = Word2Vec(tokenized_sentences_combined, vector_size=50, window=3, min_count=1, workers=4)

# Generar gráficos con la semilla
generar_grafico_red(texto_esquizofrenico, model_combined, 'Texto 1 - Baja Similitud Semántica', offset_text=20, xlim=(-150, 150), ylim=(-150, 150), zlim=(-150 ,150), seed=48)
generar_grafico_red(texto_sano, model_combined, 'Texto 2 - Alta Similitud Semántica', offset_text=20, xlim=(-150, 150), ylim=(-150, 150), zlim=(-150 ,150), seed=48)
```

```python
import numpy as np
from nltk.tokenize import sent_tokenize, word_tokenize

# Función para obtener el vector de una oración
def obtener_vector_oracion(oracion, model):
    # Tokenizar la oración en palabras
    palabras = word_tokenize(oracion, language='spanish')
    
    # Filtrar las palabras que están en el vocabulario del modelo
    vectores = [model.wv[word] for word in palabras if word in model.wv]
    
    # Calcular el vector de la oración como la suma de los vectores de las palabras
    if vectores:  # Asegurarse de que la lista no esté vacía
        vector_oracion = np.sum(vectores, axis=0)  # Cambiado a suma
        return vector_oracion
    else:
        return None  # Si no hay palabras en el vocabulario

# Texto del sujeto esquizofrénico con menor TTR
texto_esquizofrenico = 'El jardín está bien las flores están, el sol está, todo está. Las cosas están, el perro está, la mesa está No sé todo está ahí.'

# Texto del sujeto sano con mayor variedad de palabras
texto_sano = 'El jardín está hermoso. Las flores han florecido con colores brillantes, y las hojas de los árboles se mueven suavemente con la brisa. El césped está verde y el aire fresco. Se siente paz aquí.'

# Obtener los vectores de oraciones para el texto esquizofrénico
oraciones_esquizofrenico = sent_tokenize(texto_esquizofrenico, language='spanish')
vectores_oraciones_esquizofrenico = [obtener_vector_oracion(oracion, model_combined) for oracion in oraciones_esquizofrenico]

# Obtener los vectores de oraciones para el texto sano
oraciones_sano = sent_tokenize(texto_sano, language='spanish')
vectores_oraciones_sano = [obtener_vector_oracion(oracion, model_combined) for oracion in oraciones_sano]

# Mostrar los vectores de oraciones
print("Vectores de oraciones para el texto esquizofrénico:")
for oracion, vector in zip(oraciones_esquizofrenico, vectores_oraciones_esquizofrenico):
    print(f"Oración: {oracion} \nVector: {vector}\n")

print("Vectores de oraciones para el texto sano:")
for oracion, vector in zip(oraciones_sano, vectores_oraciones_sano):
    print(f"Oración: {oracion} \nVector: {vector}\n")
```

```python
import matplotlib.pyplot as plt
from collections import Counter
from math import log2
from nltk.tokenize import word_tokenize
import os

# Función para calcular la entropía iterativa
def calculate_entropy_iterative(text, language="spanish"):
    tokens = word_tokenize(text, language=language)
    entropies_original = []
    entropies_reversed = []

    # Cálculo de la entropía progresiva para el texto original
    for i in range(len(tokens)):
        partial_text = " ".join(tokens[:i+1])
        entropy = calculate_entropy(partial_text, language=language)
        entropies_original.append(entropy)

    # Cálculo de la entropía progresiva para el texto invertido
    for i in range(len(tokens)):
        partial_text_reversed = " ".join(tokens[-i-1:])
        entropy_reversed = calculate_entropy(partial_text_reversed, language=language)
        entropies_reversed.append(entropy_reversed)

    return entropies_original, entropies_reversed

# Función para calcular la entropía de un texto
def calculate_entropy(text, language="spanish"):
    tokens = word_tokenize(text, language=language)
    word_freq = Counter(tokens)
    total_words = len(tokens)
    entropy = 0

    # Cálculo de la entropía de Shannon
    for word in word_freq:
        probability = word_freq[word] / total_words
        entropy += probability * log2(probability)

    return -entropy

# Función para graficar la entropía progresiva con opción de colores personalizados y título
def plot_progressive_entropy(entropies_original, entropies_reversed, color_original='b', color_reversed='r', title="Entropía progresiva"):
    plt.figure(figsize=(4,4))
    plt.plot(range(1, len(entropies_original) + 1), entropies_original, marker='o', linestyle='-', color=color_original, label='Texto original')
    plt.plot(range(len(entropies_reversed), 0, -1), entropies_reversed, marker='o', linestyle='-', color=color_reversed, label='Texto inverso')
    plt.xlabel('Palabras')
    plt.ylabel('Media progresiva de entropía')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.ylim(0,5.5)
    plt.show()

# Textos proporcionados
texto_esquizofrenico = 'El jardín está bien las flores están, el sol está, todo está. Las cosas están, el perro está, la mesa está No sé todo está ahí.'
texto_sano = 'El jardín está hermoso. Las flores han florecido con colores brillantes, y las hojas de los árboles se mueven suavemente con la brisa. El césped está verde y el aire fresco. Se siente paz aquí.'

# Calcular entropías progresivas para ambos textos
entropias_original_esquizofrenico, entropias_reversed_esquizofrenico = calculate_entropy_iterative(texto_esquizofrenico)
entropias_original_sano, entropias_reversed_sano = calculate_entropy_iterative(texto_sano)

# Graficar con colores personalizados y título para el texto esquizofrénico
plot_progressive_entropy(entropias_original_esquizofrenico, entropias_reversed_esquizofrenico, color_original='firebrick', color_reversed='navy', title="Entropía progresiva - Texto 1")

# Graficar con colores personalizados y título para el texto sano
plot_progressive_entropy(entropias_original_sano, entropias_reversed_sano, color_original='firebrick', color_reversed='navy', title="Entropía progresiva - Texto 2")
```

```python
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize

# Función para calcular el Mean Length of Utterance (MLU)
def calcular_mlu(texto, language="spanish"):
    """
    Calcula el Mean Length of Utterance (MLU) de un texto dado.
    El MLU se calcula como el número total de palabras dividido por el número de enunciaciones.

    :param texto: Texto a analizar.
    :return: MLU del texto.
    """
    # Tokeniza el texto en enunciaciones y luego en palabras
    enunciaciones = sent_tokenize(texto, language=language)
    total_palabras = sum(len(word_tokenize(enunciacion)) for enunciacion in enunciaciones)
    mlu = total_palabras / len(enunciaciones) if enunciaciones else 0
    return mlu

# Textos proporcionados
texto_esquizofrenico = 'El jardín está bien las flores están, el sol está, todo está. Las cosas están, el perro está, la mesa está No sé todo está ahí.'
texto_sano = 'El jardín está hermoso. Las flores han florecido con colores brillantes, y las hojas de los árboles se mueven suavemente con la brisa. El césped está verde y el aire fresco. Se siente paz aquí.'

# Calcular MLU para ambos textos
mlu_esquizofrenico = calcular_mlu(texto_esquizofrenico)
mlu_sano = calcular_mlu(texto_sano)

# Graficar MLU
plt.figure(figsize=(3, 3))
textos = ['Texto 1', 'Texto 2']
mlu_values = [mlu_esquizofrenico, mlu_sano]

plt.bar(textos, mlu_values, color=['purple', 'green'])
plt.ylabel('MLU', fontsize=14)
plt.title('', fontsize=16)
#plt.ylim(0, max(mlu_values) + 1)  # Ajustar límite superior del eje Y
#plt.grid(axis='y')
plt.grid(False)
plt.tick_params(axis='both', labelsize=12)

plt.xticks(fontsize=12)
plt.show()
```

```python
import matplotlib.pyplot as plt
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

# Función para calcular la densidad de ideas (Idea Density)
def calcular_idea_density(texto, language='spanish'):
    oraciones = sent_tokenize(texto, language=language)
    palabras_vacias = set(stopwords.words(language))
    palabras_tokens = [word_tokenize(oracion, language=language) for oracion in oraciones]
    palabras_filtradas = [[palabra for palabra in palabras if palabra.lower() not in palabras_vacias] for palabras in palabras_tokens]
    conteo_ideas = sum(len(palabras) for palabras in palabras_filtradas)
    total_palabras = sum(len(palabras) for palabras in palabras_tokens)
    idea_density = conteo_ideas / total_palabras if total_palabras > 0 else 0
    return idea_density

# Textos proporcionados
texto_esquizofrenico = 'El jardín está bien las flores están, el sol está, todo está. Las cosas están, el perro está, la mesa está No sé todo está ahí.'
texto_sano = 'El jardín está hermoso. Las flores han florecido con colores brillantes, y las hojas de los árboles se mueven suavemente con la brisa. El césped está verde y el aire fresco. Se siente paz aquí.'

# Calcular Idea Density para ambos textos
idea_density_esquizofrenico = calcular_idea_density(texto_esquizofrenico)
idea_density_sano = calcular_idea_density(texto_sano)

# Graficar Idea Density
plt.figure(figsize=(3, 3))
textos = ['Texto 1', 'Texto 2']
idea_density_values = [idea_density_esquizofrenico, idea_density_sano]

plt.bar(textos, idea_density_values, color=['purple', 'green'])
plt.ylabel('Densidad de Idea', fontsize=14)
plt.title('')
#plt.ylim(0, max(idea_density_values) + 0.1)  # Ajustar límite superior del eje Y
#plt.grid(axis='y')
plt.grid(False)
plt.tick_params(axis='both', labelsize=12)

plt.xticks(fontsize=12)
plt.show()
```
